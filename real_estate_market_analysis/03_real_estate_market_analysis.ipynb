## Предобработка данных

### Работа с пропусками 

Из предыдущего пункта стало очевидно, что нам потребуется провести большую работу с пропусками. Для начала найдем долю пропусков в столбцах:

```python
df.isna().sum()/len(df)
```

Наименьшее количество пропусков в столбцах `floors_total`(всего этажей в доме) и `locality_name`(название населённого пункта). Оба этих столбца нужны для нашего исследования. Однако, заполнить их каким-либо значением не представляется возможным. Этажность у домов может быть очень разной, медиана или среднее значение не подойдет. Название населенного пункта тоже нельзя вычислить математически. Строки с этими пропусками составляют менее 1% от общего количества данных. Лучше всего их просто удалить. 

```python
df = df.dropna(subset=['floors_total', 'locality_name']).reset_index(drop=True)
```

Далее пойдем по списку с пропусками в порядке их перечисления. Так, например, в столбце `ceiling_height`(высота потолков) мы видим 9195 пропусков. Так как мы имеем дело с жилыми помещениями, пропуски можно заполнить нормативным значением по СНиП 2.08.01-89. Климатический район строительства для Санкт-Петербурга и окрестностей — IIB. Согласно СНиП 2.08.01-89, высота потолков должна быть не менее 2,5 м. Заполним пропуски в столбце значением 2,5: 

```python
df['ceiling_height'] = df['ceiling_height'].fillna(2.5) 
df.head(6) #посмотрим, как выглядят строки с индексами 1, 2, 3 - ранее там были пропуски
```

Для заполнения пропусков в столбце `living_area`(жилая площадь) сгруппируем данные по количеству комнат `rooms`, найдем медиану и заполним ею пропуски: 

```python
df = df.fillna({'living_area': df.groupby('rooms')['living_area'].transform('median')})
df.head()
```

Аналогичным способом заполним пропуски в столбце `kitchen_area`(площадь кухни):

```python
df = df.fillna({'kitchen_area': df.groupby('rooms')['kitchen_area'].transform('median')})
df.head()
```

```python
correct_area = df['kitchen_area'] + df['living_area']
np.logical_or(df['total_area'], correct_area).value_counts()
```

Рекордное количество пропусков в столбце `is_apartment`(апартаменты) можно объяснить: апартаменты достаточно редкий вид недвижимости. Кроме того, в этом столбце булев тип данных (1 или 0). Получается, что пользователи, которые не заполнили строчку "апартаменты", автоматически причислили свою недвижимость к обычной квартире. Можно заполнить пропуски в этом столбце нулями:

```python
df['is_apartment'] = df['is_apartment'].fillna('0')
df.head()
```

Аналогичная ситуация с балконами. Скорее всего, пропуски в столбце `balcony` обусловлены отсутствием балконов в квартире. Можно заполнить эти пропуски нулями:

```python
df['balcony'] = df['balcony'].fillna('0')
df.head()
```

Отдельно стоит прокомментировать невозможность заполнения пропусков в столбце `cityCenters_nearest`(расстояние до центра города). Эти данные нужны для исследования, но придется оставить их без изменений. Никаким медианным значением автоматически заполнить пропуски не получится, так как в датасете целые населённые пункты не имеют ни одного значения расстояния. Поэтому и среднее по населённому  пункту тоже будет NaN. Например, строчка кода не решит задачу:

```python
#df = df.fillna({'cityCenters_nearest': df.groupby('locality_name')['cityCenters_nearest'].transform('median')})
```

Кроме того, есть ещё 4 столбца с пропусками, которые невозможно заполнить: 
- `airports_nearest` — расстояние до ближайшего аэропорта
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма (м)

Это картографические данные, заполнять их медианным или средним значением нельзя. Кроме того, данные из этих столбцов **не будут использоваться** в нашем исследовании. Пропуски могут обозначать отсутствие парков и водоемов в радиусе 3 км от дома. В рамках этого исследования придется оставить столбцы без внимания. 

Последний столбец на очереди — `days_exposition`(сколько дней было размещено объявление). Разумно предположить, что квартиры с большим количеством комнат продаются дольше. Поэтому сгруппируем квартиры по количеству комнат, найдем медиану и заполним ею пропуски: 

```python
df = df.fillna({'days_exposition': df.groupby('rooms')['days_exposition'].transform('median')})
df.head() #обратим внимание на строчку с индексом 0, она получила значение медианы 
```

Удалим столбцы, которые не будут задействованы в исследовании. Это позволит просматривать все оставшиеся столбцы в таблице: 

```python
df_updated = df.drop(
    columns = ['parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest', 
               'airports_nearest', 'is_apartment', 'open_plan', 'studio', 'balcony', 'total_images'], axis = 1)
df_updated.head()
```

### Работа с дубликатами

Проверим таблицу на явные дубликаты: 

```python
df_updated.duplicated().sum()
```

Явных дубликатов нет. Проверим на неявные: 

```python
sorted(df_updated['locality_name'].unique())
```

Можно увидеть, что один и тот же населенный пункт называется поселком/селом/деревней. Кроме того, в глаза бросается буква ё. 

Удалим неявные дубликаты с помощью функции: 

```python
def replace_wrong_values(wrong_values, correct_value):
    for wrong_value in wrong_values: 
        df_updated['locality_name'] = df_updated['locality_name'].str.replace(wrong_value, correct_value) 

duplicates = ['посёлок', 'посёлок городского типа'] 
name = 'поселок'
replace_wrong_values(duplicates, name)
```

```python
sorted(df_updated['locality_name'].unique())
```

На этом работу с дубликатами можно считать законченной. 

### Изменение типов данных

Изменение типов данных позволяет сэкономить память и сделать данные более читаемыми. 

Наченем со столбца `last_price`(цена на момент снятия с публикации). Приведем его к целочисленному типу: 

```python
df_updated['last_price'] = df_updated['last_price'].astype('int')
df_updated.head() #проверим результат преобразования
```

Следующий кандидат на замену типа данных — `floors_total`(всего этажей в доме). Этот столбец также можно привести к целочисленному типу, т.к. чаще всего в жилых домах количество этажей обозначается целым числом. 

```python
df_updated['floors_total'] = df_updated['floors_total'].astype('int8')
df_updated.head()
```

Аналогичная ситуация со столбцом `days_exposition`. Оптимальным типом данных для него также будет челочисленный. 

```python
df_updated['days_exposition'] = df_updated['days_exposition'].astype('int16')
df_updated.head()
```

Чтобы преобразовать тип данных в столбце, содержащем пропуски, будет использовать тип `Int64`: 

```python
df_updated['cityCenters_nearest'] = df_updated['cityCenters_nearest'].astype('Int64')
df_updated.head()
```

```python
df_updated['total_area'] = df_updated['total_area'].astype('float32')
```

```python
df_updated['living_area'] = df_updated['living_area'].astype('float32')
```

```python
df_updated['kitchen_area'] = df_updated['kitchen_area'].astype('float32')
```

## Расчёты и добавление результатов в таблицу

### Расчёт цены квадратного метра

Узнаем стоимость квадратного метра. Для этого разделим стоимость объекта недвижимости на его площадь. 

Добавим столбец с ценой за квадратный метр в нашу таблицу. Сразу избавимся от цифр после запятой — в дробной части содержатся копейки, которые можно отбросить без искажения результатов исследования.

```python
df_updated['sq_price'] = (df_updated['last_price']/df_updated['total_area']).astype('int')
df_updated.head()
```

### Добавление дня недели, месяца и года публикации

Выясним и добавим в таблицу день недели, месяц и год публикации объявления: 

```python
df_updated['first_day_exposition'] = pd.to_datetime(df_updated['first_day_exposition'], format='%Y-%m-%d %H:%M')
```

```python
df_updated['weekday'] = df_updated['first_day_exposition'].dt.weekday
```

```python
df_updated['month'] = df_updated['first_day_exposition'].dt.month
```

```python
df_updated['year'] = df_updated['first_day_exposition'].dt.year
```

```python
df_updated.head()
```

### Добавление категории этажа

Проведем категоризацию этажей квартир и добавим данные в таблицу. 

Этажи будут трех категорий — первый, последний, другой.

```python
filter1 = df_updated['floor'] != 1 
filter2 = df_updated['floor'] != df_updated['floors_total']
df_updated['floor_category'] = (
    df_updated['floor'].where((filter1&filter2)==False, 'другой')
    .where(filter1, 'первый')
    .where(filter2, 'последний'))
```

```python
df_updated.head() #посмотрим, как всё прошло
```

### Добавление соотношения площадей:

В этом пункте будем искать и добавлять в таблицу соотношения различных площадей. 

Начнем с соотношения жилой и общей площади:

```python
df_updated['liv_and_total_ratio'] = (df_updated['living_area']/df_updated['total_area'])
df_updated.head()
```

Теперь найдем и добавим отношение площади кухни к общей площади: 

```python
df_updated['kitch_and_total_ratio'] = (df_updated['kitchen_area']/df_updated['total_area'])
df_updated.head()
```
